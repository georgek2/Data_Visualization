```python
# Import the libraries

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the data

data = pd.read_csv('insurance.csv')

data.head()
```

```python
# A scatter plot of  'bmi' and insurance charges
sns.scatterplot(x=data['bmi'], y=data['charges'])
```

```python
# Add a regression line 
sns.regplot(x=data['bmi'], y=data['charges'])
```

```python
# Introducing a third categorical variable to the scatterplot
plt.title('Body Mass Index vs Charges')
sns.scatterplot(x=data['bmi'], y=data['charges'], hue=data['smoker'])
```

```python
# Adding a regression line for each category of smokers
sns.lmplot(x='bmi', y='charges', hue='smoker', data=data)
```

```python
# A categorical scatterplot
sns.swarmplot(x=data['smoker'], y=data['charges'])
```

```python
# Age vs charges
# Plus categorizing smokers and non-smokers
sns.scatterplot(x=data['age'], y=data['charges'], hue=data['smoker'])
```

```python
# Charges distribution based on gender
sns.swarmplot(x=data['sex'], y=data['charges'])
```

```python
# Age distribution
sns.distplot(data['age'], kde=False)
```
