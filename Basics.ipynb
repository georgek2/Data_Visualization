```python
# Import the libraries

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the data

data = pd.read_csv('insurance.csv')

data.head()
```

```python
plt.title('Body Mass Index vs Charges')
sns.scatterplot(x=data['bmi'], y=data['charges'], hue=data['smoker'])
```

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=data)
```

```python
sns.swarmplot(x=data['smoker'], y=data['charges'])
```

```python
sns.scatterplot(x=data['age'], y=data['charges'], hue=data['smoker'])
```
